# **Filters**

```python
image_gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
image_gray[image_gray>100]=255
image_gray[image_gray<=100]=0
```

```python
cv2_imshow(image_gray)
```

```python
img = cv2.imread('Lena.png')
kernel1 = np.array([[1/9, 1/9, 1/9],
                    [1/9, 1/9, 1/9],
                    [1/9, 1/9, 1/9]])

blur = cv2.filter2D(src=img, ddepth=-1, kernel=kernel1, )
together = np.concatenate((img,blur),axis=1)
cv2_imshow(together)
```

```python
img_gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
kernel1 = np.array([[-1, 0, 1],
                    [-2, 0, 2],
                    [-1, 0, 1]])

identity = cv2.filter2D(src=img_gray, ddepth=-1, kernel=kernel1, )
cv2_imshow(identity)
```
