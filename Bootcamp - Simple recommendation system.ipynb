```python
data = data[[ 'revenue', 'title','vote_average', 'vote_count','runtime']]
```

v - no.of votes generated by the movie
c - mean rating of the perticuar data
R - mean rating of the movie
m - mini no. of votes required for the movie to be in the chart

```python
c = data['vote_average'].mean()
```

```python
m = data['vote_count'].quantile(0.9)
m
```

```python
data['runtime'].min()
```

```python
data['runtime'].max()
```

```python
filt_movies = data[(data['runtime']>=60) & (data['runtime'] <=300)]
```

```python
filt_movies = filt_movies[filt_movies['vote_count']>= m]
```

```python
filt_movies.shape
```

```python
def weighted_rating(x, m = m, c = c):
    v = x['vote_count']
    R = x['vote_average']
    wgt_rating = (v/(v+m) * R) + (m/(m+v) * c)
    return wgt_rating
```

```python
filt_movies['score']=filt_movies.apply(weighted_rating, axis=1)
```

```python
filt_movies
```

```python
filt_movies = filt_movies.sort_values("score",ascending=False)
```
